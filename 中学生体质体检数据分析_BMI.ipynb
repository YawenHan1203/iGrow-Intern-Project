```python
data_reduced=data[['BMI','bmivalue','Gender','Weight','Height','shayan','sejue','xinzang','fei','ganpi','toujing','jizhu','sizhi','pifu','xiongbu','jiemoyan','xueya','xueya_one','xueya_two','xuehong','maibo','luoyanzuoyan','luoyanyouyan','Final_Grade','FeiHuoLiang','50m','LiDingTiaoYuan','ZuoWeiTiQianQu','1km','YangWoQiZuo','YinTiXiangShang','r_yingyang','r_yingyang_bmi']]
data_reduced.head()
```

## 处理分类数据

```python
data_reduced.loc[data_reduced['BMI']=='S','BMI']=1.0
data_reduced.loc[data_reduced['BMI']=='M','BMI']=2.0
data_reduced.loc[data_reduced['BMI']=='L','BMI']=3.0
data_reduced.loc[data_reduced['BMI']=='F','BMI']=4.0

data_reduced.loc[data_reduced['Gender']=='F','Gender']=1.0
data_reduced.loc[data_reduced['Gender']=='M','Gender']=2.0
```

## 处理空值

```python
# 删除“性别”和“BMI”中空值的那一行
data_reduced=data_reduced.dropna(subset = ['Gender'])
data_reduced=data_reduced.dropna(subset = ['BMI'])
data_reduced=data_reduced.dropna(subset = ['bmivalue'])
del data_reduced['xueya_two']

#mean imputation
#axis = 0,表示列向，采用每一列的平均值填补空值
from sklearn.preprocessing import Imputer
imr = Imputer(missing_values = 'NaN', strategy = 'mean', axis = 0 )
imr = imr.fit(data_reduced.values)
imputed_data = imr.transform(data_reduced.values)
imputed_data=pd.DataFrame(imputed_data)
imputed_data.columns=[['BMI','bmivalue','Gender','Weight','Height','shayan','sejue','xinzang','fei','ganpi','toujing','jizhu','sizhi','pifu','xiongbu','jiemoyan','xueya','xueya_one','xuehong','maibo','luoyanzuoyan','luoyanyouyan','Final_Grade','FeiHuoLiang','50m','LiDingTiaoYuan','ZuoWeiTiQianQu','1km','YangWoQiZuo','YinTiXiangShang','r_yingyang','r_yingyang_bmi']]
```

## 肥胖，身高，体重，年龄综合分析

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
data_ana = data[['BMI','Height','Weight','Age']]
data_ana.head()
```

```python
#提取数据，转化为矩阵形式储存
data_ana_matrix=data_ana[['Height','Weight','Age']].as_matrix()
data_ana_assignment = data_ana['BMI'].as_matrix()

#假定认为bmi在0-50之间为正确数据,r_yingyang在960-1060之间为有效数据
colors = ['r','y','b','g']
l=['S','M','L','F']
length=len(data_ana_assignment)
fig, ax = plt.subplots(figsize=(20,15))
ax = fig.add_subplot(111, projection='3d')

for i in range(4):
    points = np.array([data_ana_matrix[j] for j in range(length) if data_ana_assignment[j]==l[i]])
    ax.scatter(xs=points[:,0],ys=points[:,1],zs=points[:,2],color=colors[i],label=l[i])
plt.legend(loc='best')
plt.xlabel('Height')
plt.ylabel('Weight')
ax.set_zlabel('Age')
ax.view_init(elev=30,azim=20)
plt.title('Scatterplot of Height, Weight and Age')
```

## PCA 降维

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit_transform(imputed_data_std.iloc[:,1:])
```

```python
explained_variance=pca.explained_variance_
ratio=explained_variance/explained_variance.sum()*100
ratio_table=pd.DataFrame(ratio)
ratio_table=ratio_table.transpose()
ratio_table.index=['Variance Ratio']
ratio_table
```

```python
covariance=pca.get_covariance() #covariance
```

```python
fig2=plt.figure(figsize=(8,6))
plt.bar(range(31), ratio, alpha=0.5, align='center', label='Individual Explained Variance')
plt.ylabel('Explained Variance Ratio Percentage')
plt.xlabel('Principal Components')
plt.legend(loc='best')
plt.tight_layout()
```
